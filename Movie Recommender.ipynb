```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_comapny
# production_countries
# release-date(not sure)
```

```python
movies.shape
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(obj):
    L = []
    counter =0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
        counter +=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])

    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
def collapse(L):
    L1=[]
    for i in L:
        L1.append(i.replace(' ',''))
    return L1
```

```python
movies['cast'] =movies['cast'].apply(collapse)
movies['genres'] =movies['genres'].apply(collapse)
movies['keywords'] =movies['keywords'].apply(collapse)
movies['crew'] =movies['crew'].apply(collapse)
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies = movies.dropna()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return ' '.join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
cv.get_feature_names_out().shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    if movie not in new['title'].values:
        print(f"Movie '{movie}' not found in the dataset.")
        return

    index = new[new['title'] == movie].index[0]

    # Check if index is valid for similarity matrix
    if index >= len(similarity):
        print(f"Index {index} is out of range for similarity matrix of shape {len(similarity)}.")
        return

    distances = list(enumerate(similarity[index]))
    distances = sorted(distances, reverse=True, key=lambda x: x[1])

    print(f"\nRecommendations for '{movie}':")
    for i in distances[1:6]:  # Skip the first (which is the movie itself)
        print(new.iloc[i[0]].title)
```

```python
recommend('John Carter')
```
